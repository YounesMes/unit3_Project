```python
true_label=[]
for i in df['Category']:
    true_label.append(i)
```

```python
for j in df['Predicted_class']:
    predict_class.append(j)
```

```python
for values in range(0,len(cluster)):
    if cluster[values]==0: 
        true_class0.append(true_label[values])
        predict_class0.append(predict_class[values])
    if cluster[values]==1: 
        true_class1.append(true_label[values])
        predict_class1.append(predict_class[values])
    if cluster[values]==2:
        true_class2.append(true_label[values])
        predict_class2.append(predict_class[values])
    if cluster[values]==3:
        true_class3.append(true_label[values])
        predict_class3.append(predict_class[values])
    if cluster[values]==4:
        true_class4.append(true_label[values])
        predict_class4.append(predict_class[values])
```
